```python
df1=df[df['Date_N'].dt.year==2019]
Releases_2019=df1['Date_N'].value_counts()
```

```python
Releases_2019
```

```python
Releases_2019.sort_index(inplace=True)
```

```python
Releases_2019.plot()
import matplotlib.pyplot as plt
plt.title('Releases in the year 2019')
plt.ylabel('Total number of movies and shows')
plt.xlabel('Date')
```

```python
df2=df[df['Date_N'].dt.year==2021]
Releases_2021=df2['Date_N'].value_counts()
```

```python
Releases_2021.sort_index(inplace=True)
```

```python
Releases_2021.plot()
import matplotlib.pyplot as plt
plt.title('Releases in the year 2021')
plt.ylabel('Total number of movies and shows')
plt.xlabel('Date')
```

```python
category_2021=df2.groupby(['Date_N','Category'])['Date_N'].count().unstack()
```

```python
category_2021.plot()
plt.title('Releases in the year 2021')
plt.ylabel('Total number of movies and shows')
plt.xlabel('Date')
```
